```python
from Fuzzy_set import Fuzzy_set, Fuzzy_field
```

### This example is presented to show the capability of the library.

### Suppose we need to calculate the profitability of a company based on its three sources of funding and 1 debt.

```python
A = Fuzzy_set(50,70,10,30) # Source of financing
B = Fuzzy_set(100,110,0,0) # Source of financing
C = Fuzzy_set(0,0,0,20)    # Source of financing
D = Fuzzy_set(80,80,30,20) # Company debt

'''
Fund of the company, taking into account sources of financing and excluding debt
'''
F = A + B + C

'''
Debt-adjusted company fund
'''
FD = F - D

ROI = FD / D # Company profitability
```

## Visualization of funding sources

```python
Q = Fuzzy_field(A, B, C)
Q.draw_field(title='Funding sources A, B, C')
```

## Visualization of сompany debt

```python
D.draw_set(title='Debt B')
```

### Visualization of the company's fund without taking into account debt and taking into account debt

```python
Q.clear()
Q.add(F, FD)
Q.draw_field(title='Foundation F and FD')
print(f'{F = }', f'{FD = }', sep='\n')
```

## Visualization of projected profitability of the company

```python
ROI.draw_set(title='ROI')
```

### Let's find the lower and upper intersection of one source of financing A and the fund of the company FD

```python
print('Supremum:', A.supr(FD))
print('Infimum:', A.inf(~FD))
```
