# Algebraic operation of fuzzy sets

### Initialize 3 sets and output them

```python
A = Fuzzy_set(3, 5, 2, 2)
B = Fuzzy_set(7, 9, 2, 2)
C = Fuzzy_set(6, 6, 2, 2)
Q = Fuzzy_field(A,B,C)
Q.draw_field()
```

## Truncation of fuzzy set

$$ \mu_{k\cap A}(x) = \min\{k,\ \mu_{A}(x)\} $$

```python
A.trunc(0.5)
```

## Extension of fuzzy set

$$ \mu_{k\cap A}(x) = \max\{k,\ \mu_{A}(x)\} $$

```python
A.extension(0.5)
```

## Concentration of fuzzy set

$$ \mu_{Con(A)}(x) = (\mu_{A}(x))^\beta,\ \beta > 1$$

```python
A.con(1.5)
```

## Dilatation of fuzzy set

$$ \mu_{Dil(A)}(x) = (\mu_{A}(x))^\beta,\ 0 < \beta < 1$$

```python
A.dil(0.5)
```

## Equivalence of 2 fuzzy sets

$$ \mu_{A\ \equiv\ B}(x) = 1 - |\mu_{A}(x) - \mu_{B}(x)|$$

```python
A == B
```

## Not equivalence of 2 Fuzzy Sets

$$ \mu_{A\ \not\equiv\ B}(x) = 1 - \mu_{A\ \equiv\ B}(x) = |\mu_{A}(x) - \mu_{B}(x)|$$

```python
A != B
```

## Inclusion of one fuzzy set into another

$$ \mu_{A\ 	\subseteq\ B}(x) = \min\{1,\ 1 - \mu_{A}(x) + \mu_{B}(x)\} $$

```python
A.inclusion(B)
```

## Algebraic addition of 2 fuzzy sets

$$ \mu_{A + B}(x) = \mu_{A}(x) + \mu_{B}(x) - \mu_{A}(x)\cdot\mu_{B}(x)$$

```python
A.addition(B)
```

## Algebraic substraction of 2 or more fuzzy sets

$$ \mu_{A-B}(x) = \max(0,\ \mu_{A}(x) - \mu_{B}(x) - \ldots)$$

```python
A.substract(C)
```

## Algebraic multiplication of 2 or more fuzzy sets

$$ \mu_{A\cdot B}(x) = \mu_{A}(x)\cdot \mu_{B}(x)\cdot\ldots $$

```python
A.mul(B)
```

## Multiplication of a fuzzy set by a constant

$$ \mu_{k\cdot A}(x) = \min\{1,\ k\cdot \mu_{A}(x)\}$$

```python
A.mul(0.5)
```

```python
A.mul(2)
```

## Union of 2 or more fuzzy sets

$$ \mu_{A\cup B}(x) = \max\{\mu_{A}(x),\ \mu_{B}(x), \ldots\} $$

```python
A.union(B)
```

## intersection of 2 or more fuzzy sets

$$ \mu_{A\cap B}(x) = \min\{\mu_{A}(x),\ \mu_{B}(x), \ldots\} $$

```python
A.intersect(B)
```
